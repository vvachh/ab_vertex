```python
import numpy as np
import matplotlib.pyplot as plt
from vertex_model_ab import VertexModelApicoBasal
from vertex_model_2d import VertexModel
from voronoi_cells import random_voronoi_cells
```

```python
locs,adjs = random_voronoi_cells(4)
```

```python
vm2d = VertexModel(locs,adjs,boxsize=4,p0=1.,r=0.1)
vm2d.show_model(); plt.show()
print 'Initial energy:',vm2d.total_mechanical_energy()
print 'rosettes:',vm2d.find_rosettes(0.2)
vm2d.gradient_descent_with_t1(0.00001,0.01,0.2)
vm2d.show_model(); plt.show()
print 'Final energy:',vm2d.total_mechanical_energy()
print 'rosettes:',vm2d.find_rosettes(0.2)
```

```python
vmab = VertexModelApicoBasal(locs,adjs,0.0,boxsizeb=4,h=1.,r=0.1,A0=4.81)
vmab.show_model()
print 'Initial energy:',vmab.combined_energy()
ap_ros,bas_ros,shared_ros = vmab.classify_rosettes(0.1)
print 'apical rosettes:',ap_ros
print 'basal rosettes:',bas_ros
print 'shared rosettes:',shared_ros
vmab.gradient_descent_with_t1(0.00001,.01,0.1)
plt.figure()
vmab.show_model()
print 'Final energy:',vmab.combined_energy()
ap_ros,bas_ros,shared_ros = vmab.classify_rosettes(0.1)
print 'apical rosettes:',ap_ros
print 'basal rosettes:',bas_ros
print 'shared rosettes:',shared_ros
```
